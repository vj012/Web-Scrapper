```python
df = pd.read_excel (r'Amazon Scraping.xlsx')
with open("output.json", "w") as outfile:
  for k in range(1000):
    user_agent = random.choice(user_agent_list) 
    headers = {'User-Agent': user_agent}
    sleep(0.01*randint(10,100))
    start = time.time()
    my_dict = {}
    product_description={}
    asin=df.Asin[k]
    country=df.country[k]
    print(k,country,asin)
    url= f"https://www.amazon.{country}/dp/{asin}"
    html_text = requests.get(url,headers=headers)
    status = html_text.status_code
    if status == 404:
        my_dict[html_text.url]='not avaliable'
    else:
        soup = BeautifulSoup(html_text.text, 'lxml')
        product_title = soup.find('span', id='productTitle')
        if product_title != None:
          my_dict['product title']=product_title.text.strip()

        product_imgurl = soup.find('div', id='img-canvas')
        
        if product_imgurl != None:
          img_url = product_imgurl.find('img')
          my_dict['product image url']=img_url['src']
      
        product_price = soup.find('span', class_='a-size-base a-color-price a-color-price')
        if product_price!= None:
          string = product_price.text.replace(u'\xa0', u' ')
          string=string.encode("ascii", "ignore")
          string=string.decode()
          my_dict['product price']=string.strip()
        
        product_details = soup.find('div', id='detailBullets_feature_div')
        if product_details!= None:
          x=product_details.find_all('span')
          for i in range(0,12,3):
            str1=f"{x[i+1].text.replace(' ','')}"
            str3=str1.replace('\n','')
            str3=str3.encode("ascii", "ignore")
            str3=str3.decode()
            str2=f"{x[i+2].text.replace(' ','')}"
            product_description[str3]=str2
            my_dict['product description']=product_description

    
    if k%100==0:
      end = time.time()
      my_dict['Time for execution of program for 100 order of computations : ']= round(end-start, 10)
      sleep(randint(10,100))
      
    json_object = json.dumps(my_dict, indent = 4)
    outfile.write(json_object)        
    
```
